1. A partir del siguiente código, intenta obtener el gráfico del dibujo llamado ejercicio 1 en el enunciado.

```python
n = 256
X = np.linspace(-np.pi, np.pi, n)
Y = np.sin(2 * X)

plt.figure(figsize=(10, 8))
plt.plot(X, Y + 1, color = 'blue', alpha = 1.00)
plt.plot(X, Y - 1, color = 'blue', alpha = 1.00)

ax = plt.gca()
ax.fill_between(X, Y + 1, 1, color = 'blue', alpha = 0.3)
ax.fill_between(X, Y - 1, -1, where = (Y - 1 >= -1), color = 'blue', alpha = 0.3)
ax.fill_between(X, Y - 1, -1, where = (Y - 1 < -1), color = 'red', alpha = 0.3)

plt.show()
```

Respuesta profesor

```python
plt.axes([0.025, 0.025, 0.95, 0.95])

plt.plot(X, Y + 1, color='blue', alpha=1.00)

plt.fill_between(X, 1, Y + 1, color='blue', alpha=.25)


plt.plot(X, Y - 1, color='blue', alpha=1.00)

plt.fill_between(X, -1, Y - 1, (Y - 1) > -1, color='blue', alpha=.25)

plt.fill_between(X, -1, Y - 1, (Y - 1) < -1, color='red',  alpha=.25)


plt.xlim(-np.pi, np.pi)

plt.xticks(())

plt.ylim(-2.5, 2.5)

plt.yticks(())


plt.show()
```

```python
n = 1024
X = np.random.normal(0, 1, n)
Y = np.random.normal(0, 1, n)

#calculamos los angulos
ang = np.arctan2(Y, X)
#seteamos la variable colors (solo para que se entienda más)
colors = ang

plt.figure(figsize = (10, 8))
plt.scatter(X,Y, c = colors, cmap = 'hsv', alpha = 0.6)
plt.show()
```

Respuesta profesor

```python
T = np.arctan2(Y, X)

plt.axes([0.025, 0.025, 0.95, 0.95])

plt.scatter(X, Y, s=75, c=T, alpha=.5)

plt.xlim(-1.5, 1.5)

plt.xticks(())

plt.ylim(-1.5, 1.5)

plt.yticks(())

plt.show()
```

```python
n = 12
X = np.arange(n)
Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)

plt.figure(figsize=(10,8))
plt.bar(X, +Y1, facecolor = '#9999ff', edgecolor = 'white')
plt.bar(X, -Y2, facecolor = '#ff9999', edgecolor = 'white')

plt.ylim((-Y2).min()*1.3, Y1.max()*1.3)

for x, y in zip(X, Y1):
    plt.text(x, y + 0.05, '%.2f' % y, ha = 'center', va = 'bottom', fontsize = 15)

for x, y in zip(X, -Y2):
    plt.text(x, y - 0.05, '%.2f' % abs(y), ha = 'center', va = 'top', fontsize = 15)
    
plt.show()
```

4. A partir del siguiente código, intenta obtener el gráfico del dibujo llamado ejercicio 4 en el enunciado.

Pista: tendrás que modificar la Z.

Z = np.random.uniform(0, 1, 20)

plt.pie(Z)

```python
Z = np.random.uniform(0, 1, 20)
Z.sort()
Z = Z[::-1]
explode = np.zeros(20)
explode[0] = 0.1
explode[1:] = 0.05

plt.figure(figsize = (10, 8))
ax = plt.gca()
theme = plt.get_cmap('Greys')
ax.set_prop_cycle("color", [theme(1. * i / len(Z))
                             for i in range(len(Z))])
plt.pie(Z, startangle=0, counterclock=False, wedgeprops=dict(linewidth=1.2, edgecolor='grey'), explode=explode)

plt.show()
```

Respuesta profesor

```python
n = 20

Z = np.ones(n)

Z[-1] *= 2

plt.axes([0.025, 0.025, 0.95, 0.95])

plt.pie(Z, explode=Z*.05, colors = ['%f' % (i/float(n)) for i in range(n)])

plt.axis('equal')

plt.xticks(())

plt.yticks()

plt.show()
```
